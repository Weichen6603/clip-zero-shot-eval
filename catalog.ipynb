```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 设置显示参数
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)

# 设置路径
csv_path = "./data/treeoflife/metadata/catalog.csv"
output_dir = "catalog_analysis_output"
os.makedirs(output_dir, exist_ok=True)

# 加载数据
df = pd.read_csv(csv_path)
print(f"✅ 数据加载完成，样本总数: {len(df)}，字段数: {df.shape[1]}")
df.head()
```

```python
with open("data/treeoflife/metadata/catalog.csv", "r", encoding="utf-8") as f:
    first_line = f.readline()
    print(repr(first_line))
```

```python
split_counts = df['split'].value_counts()
display(split_counts)

# 可视化
split_counts.plot(kind='bar', color='skyblue')
plt.title('Data Split Distribution')
plt.xlabel('Split')
plt.ylabel('Number of Samples')
plt.tight_layout()
plt.show()
```

```python
top_species = df['species'].value_counts().head(20)
display(top_species)

# 可视化
top_species.plot(kind='barh', color='orange')
plt.title('Top 20 Species')
plt.xlabel('Number of Samples')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
